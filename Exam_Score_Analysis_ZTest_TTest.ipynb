🚀***Import Libraries***

```python
# Import Required Libraries

import numpy as np
import pandas as pd

from scipy.stats import ttest_1samp, ttest_ind, ttest_rel
from statsmodels.stats.weightstats import ztest

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

📌 ***Understanding Hypothesis Testing***

**Hypothesis Testing is a statistical method used to make decisions about data.**

**It helps determine whether the observed difference is statistically significant or happened due to random chance.**



📌***Key Terms***

**Null Hypothesis (H₀)**

***The default assumption that there is no effect or difference.***

📌**Alternative Hypothesis (H₁)**

***The assumption that there is an effect or significant difference.***

📌 **Significance Level (α)**
***The threshold used to decide whether to reject the null hypothesis.***

Commonly:
***α = 0.05***

📌***p-value***
**The probability of obtaining the observed result assuming the null hypothesis is true.**

- p-value < 0.05 → Reject H₀
- p-value ≥ 0.05 → Fail to Reject H₀

📊 ***Student Exam Score Dataset***

```python
# General student scores
scores = [78, 82, 85, 88, 90, 76, 84, 79, 81, 87]

# Class datasets
class_A = [78, 82, 85, 88, 90]
class_B = [70, 72, 75, 68, 74]

# Before vs After training
before_training = [65, 70, 68, 72, 66]
after_training = [72, 75, 74, 78, 73]

print("Datasets Loaded Successfully")
```

📌 ***Understanding p-value***

**The p-value helps determine whether the result obtained from the sample data is statistically significant.**

**It measures the probability of observing the given result assuming the Null Hypothesis (H₀) is true.**



📌***Interpretation of p-value***

- p-value < 0.05 → Reject Null Hypothesis
- p-value ≥ 0.05 → Fail to Reject Null Hypothesis

**A smaller p-value indicates stronger evidence against the null hypothesis.**

📊 ***Z-Test***

📌***Problem Statement***

*A school claims that the average exam score of students is 80.*

*We will perform a Z-Test to check whether the sample mean is significantly different from 80.*

```python
# Perform Z-Test

z_stat, p_value = ztest(scores, value=80)

print("Z-Statistic :", z_stat)
print("P-Value :", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

📌 ***Observation***

**The Z-Test compares the sample mean with the hypothesized population mean.**

Based on the p-value:
- ***If p-value is less than 0.05, the average score is significantly different from 80.***
- ***Otherwise, there is not enough evidence to reject the school’s claim.***

📊***Single Sample t-Test***

📌***Problem Statement***

*A teacher claims that the average student score is 80.*

*We will perform a Single Sample t-Test to check whether the sample mean differs significantly from 80.*

```python
# Perform Single Sample t-Test

t_stat, p_value = ttest_1samp(scores, 80)

print("T-Statistic :", t_stat)
print("P-Value :", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

📌 ***Observation***

**The Single Sample t-Test is used when:**
- **population standard deviation is unknown**
- **sample size is small**

This test helps determine whether the average student score significantly differs from the claimed mean value.

📊***Independent 2-Sample t-Test***

📌***Problem Statement***

*We want to compare the performance of Class A and Class B students.*

*The Independent t-Test checks whether the mean scores of the two independent groups are significantly different.*

```python
# Perform Independent 2-Sample t-Test

t_stat, p_value = ttest_ind(class_A, class_B)

print("T-Statistic :", t_stat)
print("P-Value :", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

📌***Observation***

**The Independent t-Test compares the means of two different groups.**

Based on the p-value:
- **If p-value < 0.05 → the performance difference between the two classes is statistically significant.**
- **Otherwise, the difference may have occurred due to random variation.**

📊 ***Paired 2-Sample t-Test***

📌***Problem Statement***

**A training program was conducted to improve student performance.**

**We will compare student scores before and after training using a Paired t-Test to determine whether the training had a significant impact.**

```python
# Perform Paired 2-Sample t-Test

t_stat, p_value = ttest_rel(before_training, after_training)

print("T-Statistic :", t_stat)
print("P-Value :", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

📌 ***Observation***

The Paired t-Test is used when:
- **the same group is tested twice,**
- **such as before and after training.**

This test helps determine whether the training program significantly improved student performance.
